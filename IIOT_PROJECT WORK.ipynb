```python
# Purchased column fix
df['Purchased'] = df['Purchased'].replace({
    'Low':0, 'High':1,
    'No':0, 'Yes':1
})

df['Purchased'] = pd.to_numeric(df['Purchased'], errors='coerce')
df['Purchased'] = df['Purchased'].fillna(0)

# Numeric columns clean
cols = ['Order_Value','Time_Spent_Minutes','Items_Added_to_Cart',
        'Pages_Viewed','Review_Rating','Loyalty_Score']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    df[col] = df[col].fillna(df[col].mean())

print("Data Ready (No Errors)")
```

```python
# Loyalty Score vs Order Value
plt.figure(figsize=(8,5))
# Scatter plot
plt.scatter(
    df['Loyalty_Score'],
    df['Order_Value'],
    color='green',
    alpha=0.6
)
# Trend line
z = np.polyfit(df['Loyalty_Score'], df['Order_Value'], 1)
p = np.poly1d(z)

plt.plot(
    df['Loyalty_Score'],
    p(df['Loyalty_Score']),
    color='red',
    linewidth=2
)
# Titles and labels
plt.title('Loyalty Score vs Order Value')
plt.xlabel('Loyalty Score')
plt.ylabel('Order Value')
plt.grid(True)
plt.show()
```

```python
# Time Spent vs Order Value (Trend Analysis)
plt.figure(figsize=(8,5))
# Scatter plot
plt.scatter(
    df["Time_Spent_Minutes"],
    df["Order_Value"],
    color="purple",
    alpha=0.6
)
# Trend line (best practice)
z = np.polyfit(df["Time_Spent_Minutes"], df["Order_Value"], 1)
p = np.poly1d(z)

plt.plot(
    df["Time_Spent_Minutes"],
    p(df["Time_Spent_Minutes"]),
    color="black",
    linewidth=2
)
# Labels and title
plt.title("Time Spent vs Order Value (Trend Analysis)")
plt.xlabel("Time Spent (Minutes)")
plt.ylabel("Order Value")
plt.grid(True)
plt.show()
```

```python
# Cart Items vs Order Value (Conversion-wise)
plt.figure(figsize=(8,5))
# Split data
p0 = df[df['Purchased'] == 0]
p1 = df[df['Purchased'] == 1]
# Scatter plots with 2 colors
plt.scatter(
    p0['Items_Added_to_Cart'],
    p0['Order_Value'],
    color='blue',
    alpha=0.6,
    label='Not Purchased'
)
plt.scatter(
    p1['Items_Added_to_Cart'],
    p1['Order_Value'],
    color='red',
    alpha=0.6,
    label='Purchased'
)
# Labels and title
plt.title('Cart Items vs Order Value (Conversion-wise)')
plt.xlabel('Items Added to Cart')
plt.ylabel('Order Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Customer Behavior Analysis: Order Value, Discounts, Age & Purchase Patterns
# Create 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Customer Behavior Analysis: Order Value, Discounts, Age & Purchase Patterns', fontsize=16)
# 1. Weekend vs Weekday Order Value
sns.boxplot(
    x='Is_Weekend', 
    y='Order_Value', 
    data=df, 
    hue='Is_Weekend', 
    palette='YlGnBu', 
    ax=axes[0, 0]
)
axes[0, 0].set_title('Weekend vs Weekday Order Value')

# 2. Discount vs Order Value
sns.boxplot(
    x='Discount_Applied', 
    y='Order_Value', 
    data=df, 
    hue='Discount_Applied', 
    palette='Set1', 
    ax=axes[0, 1]
)
axes[0, 1].set_title('Discount vs Order Value')

# 3. Age Distribution across Locations
sns.boxplot(
    x='Location', 
    y='Age', 
    data=df, 
    hue='Location', 
    palette='Set3', 
    ax=axes[1, 0]
)
axes[1, 0].set_title('Age Distribution across Locations')
axes[1, 0].tick_params(axis='x')

# 4. Cart Items vs Purchase
sns.boxplot(
    x='Purchased', 
    y='Items_Added_to_Cart', 
    data=df, 
    hue='Purchased', 
    palette='coolwarm', 
    ax=axes[1, 1]
)
axes[1, 1].set_title('Cart Items vs Purchase')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show all plots
plt.show()
```
